```python
#Import pandas, numpy, and matplotlib.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Create an array with values from -10 to +10
x = np.arange(-10,11)
```

```python
#Print the array x
x
```

```python
plt.figure(figsize = (12,6))
plt.title('My Nice Plot')

plt.plot(x, x ** 2)
plt.plot(x, -1 * (x ** 2))
```

```python
plt.figure(figsize=(12, 6))
plt.title('My Nice Plot')

plt.subplot(1, 2, 1)  # It's going to have one row, two columns, and will start drawing from the first plot.
plt.plot(x, x ** 2)
plt.plot([0, 0, 0], [-10, 0, 100])
plt.legend(['X^2', 'Vertical Line'])
plt.xlabel('X')
plt.ylabel('X Squared')

plt.subplot(1, 2, 2)
plt.plot(x, -1 * (x ** 2))
plt.plot([-10, 0, 10], [-50, -50, -50])
plt.legend(['-X^2', 'Horizontal Line'])

plt.xlabel('X')
plt.ylabel('X Squared')
```

```python
fig, axes = plt.subplots(figsize = (12,6))

axes.plot(x, x + 0, linestyle = 'solid')
axes.plot(x, x + 1, linestyle = 'dashed')
axes.plot(x, x + 2, linestyle = 'dashdot')
axes.plot(x, x + 3, linestyle = 'dotted')

axes.set_title('My Nice plot')
```

```python
fig, axes = plt.subplots(figsize = (12,6))

axes.plot(x, x + 0, '-g', label = 'solid green')
axes.plot(x, x + 1, '--c',label = 'dashed cayen')
axes.plot(x, x + 2, '-.b', label = 'dashdot blue')
axes.plot(x, x + 3, ':r', label = 'dotted red')

axes.set_title('My Nice plot')

axes.legend()
```

```python
axes.plot(
    x, (x ** 2), color='red', linewidth=3,
    marker='o', markersize=8, label='X^2')

axes.plot(x, -1 * (x ** 2), 'b--', label='-X^2')

axes.set_xlabel('X')
axes.set_ylabel('X Squared')

axes.set_title("My Nice Plot")

axes.legend()
fig
```

```python
N = 50
X = np.random.randn(N)
Y = np.random.randn(N)
colors = np.random.randn(N)
area = np.pi * (20 * np.random.randn(N)) ** 2
```

```python
plt.figure(figsize=(14, 6))

plt.scatter(X, Y, s=area, c=colors, alpha=0.5, cmap='Spectral')
plt.colorbar()

plt.show()
```

```python
fig = plt.figure(figsize = (14,6))

ax1 = fig.add_subplot(1,2,1)
plt.scatter(X,Y, c=colors, s=area,alpha = 0.5, cmap = "Pastel1")
plt.colorbar()

ax2 = fig.add_subplot(1,2,2)
plt.scatter(X,Y, c= colors, s= area, alpha = 0.5, cmap = "Pastel2")
plt.colorbar()

plt.show()
```

```python
values = np.random.randn(1000)
```

```python
plt.subplots(figsize = (12,6))
plt.hist(values, bins = 100, histtype = 'bar', alpha = 0.8, color = 'steelblue', edgecolor = 'green')
plt.xlim(xmin = -5, xmax = 5)
plt.show()
```

```python
fig.savefig('hist.png')
```

```python
#Kernel Density Estimation
from scipy import stats

density = stats.kde.gaussian_kde(values)
density
```

```python
plt.subplots(figsize=(12, 6))

values2 = np.linspace(min(values)-10, max(values)+10, 100)

plt.plot(values2, density(values2), color='#FF7F00')
plt.fill_between(values2, 0, density(values2), alpha=0.5, color='#FF7F00')
plt.xlim(xmin=-5, xmax=5)

plt.show()
```

```python
#Combine plots
plt.subplots(figsize = (12,6))

plt.hist(values, bins = 100, histtype = 'bar', alpha = 0.8, density = 1, color = 'steelblue', edgecolor = 'green')
plt.xlim(xmin = -5, xmax = 5)

plt.plot(values2, density(values2), color='#FF7F00', linewidth = 3)
plt.xlim(xmin=-5, xmax=5)

plt.show()
```

```python
Y = np.random.rand(1, 5)[0]
Y2 = np.random.rand(1, 5)[0]
```

```python
plt.figure(figsize = (12,4))

barWidth = 0.5
plt.bar(np.arange(len(Y)), Y, width = barWidth, color='#00b894')

plt.show()
```

```python
plt.figure(figsize = (12,4))

plt.bar(np.arange(len(Y)), Y, width = 0.5, color='#00b894')

plt.show()
```

```python
#Stacked bars(add a legend)
plt.figure(figsize = (12,4))

plt.bar(np.arange(len(Y)), Y, width = 0.5, color='#00b894')
plt.bar(np.arange(len(Y2)), Y2, width = 0.5, color='#e17055', bottom = Y, label = Y2 )

plt.legend()
plt.show()
```
